```python
# Read in the corpus and construct the corpus DataFrame
rmrb = {"year": [], "words": []}
for file in os.listdir("../corpus_by_year/"):
    if file.endswith(".txt"):
        year = file.split(".")[0]
#         print(year)
        tokens = []
        path = "../corpus_by_year/" + file
        with open(path) as f:
            for line in f.readlines():
                tokens.append(line.strip().split())
        words = list(itertools.chain.from_iterable(tokens.copy()))
        rmrb["year"].append(year)
        rmrb["words"].append([words])
```

```python
df = pd.DataFrame(rmrb)
tokens = []
for index, year in df.iterrows():
    tokens.append(year["words"][0])
df["tokens"] = tokens
df = df[["year", "tokens"]]
df.sort_values("year")
```

## Counting words

```python
# Top n words each year
n = 20
counters = {"year":[], "top_n_words": []}
for index, year in df.iterrows():
    count = Counter(year["tokens"])
    sorted_count = sorted(count.items(), key=lambda x: x[1], reverse=True)
    counters["year"].append(year["year"])
    counters["top_n_words"].append([pair for pair in sorted_count])

# Construct dictionary of counted words by year
dict_top_n_words = {}
for idx, year in enumerate(counters["year"]):
    dict_top_n_words[year] = counters["top_n_words"][idx]
dict_top_n_words
```

```python
# Top n words each year
n = 10
top_words = {}
for k, v in dict_top_n_words.items():
    words = [pair[0] for pair in v[:n + 1]]
    top_words[k] = words
top_words_df = pd.DataFrame(top_words)
top_words_df.reindex(sorted(top_words_df.columns), axis=1)
```
